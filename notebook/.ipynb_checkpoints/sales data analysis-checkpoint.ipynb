```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import warnings
import plotly.express as px
import panel as pn
pn.extension('plotly')
import datetime
```

```python
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv("../data/sales.csv")
df.head()

```

```python
df.duplicated()
```

```python
df.info()
df.isnull().sum()
df.drop_duplicates(inplace=True)
df["Order Date"] = pd.to_datetime(df["Order Date"])
df.drop(columns=['Unnamed: 21'],inplace=True)
df.info()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")
df["Profit"] = pd.to_numeric(df["Profit"], errors="coerce")
df['Category'] = df['Category'].str.strip()
df['Sub-Category'] = df['Sub-Category'].str.strip()
df['Region'] = df['Region'].str.strip()

df = df.dropna(subset=['Order Date'])
df = df.dropna(subset=['Sales','Profit'])
df['Month'] = df['Order Date'].dt.to_period('M').dt.to_timestamp()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
plt.figure(figsize=(8,6))
barplot = sns.barplot(
    data=category_sales,
    x='Category',
    y='Sales',
    palette='Set2'
)
plt.title("Total Sales by Category", fontsize=16, weight='bold')
for index, row in category_sales.iterrows():
    barplot.text(
        index, 
        row['Sales'] + 20000, 
        f"{int(row['Sales']):,}", 
        color='black',
        ha="center",
        fontsize=12,
        weight='bold'
    )

plt.ylabel("Total Sales", fontsize=12)
plt.xlabel("Category", fontsize=12)
plt.show()
```

```python
sub_sales = df.groupby("Sub-Category")["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False)

plt.figure(figsize=(12,6))
bar = sns.barplot(data=sub_sales, x="Sub-Category", y="Sales", palette="coolwarm")
plt.title("Total Sales by Sub-Category", fontsize=16, weight="bold")
plt.xticks(rotation=45)

for i, row in sub_sales.iterrows():
    bar.text(i, row["Sales"]+2000, f"{int(row['Sales']):,}", 
             ha="center", fontsize=10, weight="bold")

plt.ylabel("Total Sales")
plt.show()

```

```python
category_profit = df.groupby("Category")["Profit"].sum().reset_index()

plt.figure(figsize=(6,3))
bar = sns.barplot(data=category_profit, x="Category", y="Profit", palette="viridis")
plt.title("Total Profit by Category", fontsize=16, weight="bold")

for i, row in category_profit.iterrows():
    bar.text(i, row["Profit"]+2000, f"{int(row['Profit']):,}", 
             ha="center", fontsize=12, weight="bold")

plt.ylabel("Total Profit")
plt.show()
```

```python
sub_profit = df.groupby("Sub-Category")["Profit"].sum().reset_index().sort_values(by="Profit", ascending=False)

plt.figure(figsize=(12,6))
bar = sns.barplot(data=sub_profit, x="Sub-Category", y="Profit", palette="crest")
plt.title("Total Profit by Sub-Category", fontsize=16, weight="bold")
plt.xticks(rotation=45)

for i, row in sub_profit.iterrows():
    bar.text(i, row["Profit"]+500, f"{int(row['Profit']):,}", 
             ha="center", fontsize=10, weight="bold")

plt.ylabel("Total Profit")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().reset_index()

plt.figure(figsize=(6,3))
bar = sns.barplot(data=region_sales, x="Region", y="Sales", palette="mako")
plt.title("Total Sales by Region", fontsize=16, weight="bold")

for i, row in region_sales.iterrows():
    bar.text(i, row["Sales"]+2000, f"{int(row['Sales']):,}", 
             ha="center", fontsize=12, weight="bold")

plt.ylabel("Total Sales")
plt.show()
```

```python
region_profit = df.groupby("Region")["Profit"].sum().reset_index()

plt.figure(figsize=(6,3))
bar = sns.barplot(data=region_profit, x="Region", y="Profit", palette="flare")
plt.title("Total Profit by Region", fontsize=16, weight="bold")

for i, row in region_profit.iterrows():
    bar.text(i, row["Profit"]+500, f"{int(row['Profit']):,}", 
             ha="center", fontsize=12, weight="bold")

plt.ylabel("Total Profit")
plt.show()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum().reset_index()
monthly_sales["Order Date"] = monthly_sales["Order Date"].dt.to_timestamp()

plt.figure(figsize=(6,3))
sns.lineplot(data=monthly_sales, x="Order Date", y="Sales", marker="o", color="teal")
plt.title("Monthly Sales Trend", fontsize=16, weight="bold")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
monthly_sales["Moving Avg (3M)"] = monthly_sales["Sales"].rolling(3).mean()

plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales, x="Order Date", y="Sales", marker="o", label="Sales")
sns.lineplot(data=monthly_sales, x="Order Date", y="Moving Avg (3M)", color="red", linewidth=2, label="3-Month Moving Avg")
plt.title("Time Series Analysis: Monthly Sales with Moving Average", fontsize=16, weight="bold")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
monthly_sales = monthly_sales.to_timestamp()
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker="o", color="teal", linewidth=2)
plt.title("Monthly Sales Trend", fontsize=16, weight="bold")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
for x, y in zip(monthly_sales.index, monthly_sales.values):
    plt.text(x, y+2000, f"{int(y):,}", ha="center", fontsize=8, rotation=45)

plt.show()
```

```python
category_summary = df.groupby("Category")[["Sales", "Profit"]].sum()
category_summary["Profit Margin %"] = (category_summary["Profit"] / category_summary["Sales"]) * 100
category_summary
```

```python
decomposition = seasonal_decompose(monthly_sales, model="additive")
fig = decomposition.plot()
fig.set_size_inches(12, 8)  
plt.suptitle("Additive Seasonal Decomposition of Monthly Sales", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
model = ARIMA(monthly_sales, order=(1, 1, 1))
results = model.fit()
forecast = results.forecast(steps=12)
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label="Historical", marker='o')
plt.plot(forecast, label="Forecast", color="red", marker='x')
plt.title("ARIMA Forecast - Next 12 Months", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
prophet_df = monthly_sales.reset_index()
prophet_df.columns = ["ds", "y"] 
model = Prophet()
model.fit(prophet_df)
future = model.make_future_dataframe(periods=12, freq="ME")  # 'ME' = Month End
forecast = model.predict(future)
fig = model.plot(forecast)
plt.title("Prophet Forecast - Next 12 Months", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
bg_color = '#1E1E1E'
card_bg_color = '#2C2C2C'
text_color = '#FFFFFF'

# --- Widget Styling ---
widget_style = {
    'background-color': card_bg_color,
    'color': text_color,
    'border': '1px solid #444444'
}

category_selector = pn.widgets.Select(
    name='Category',
    options=['All'] + df['Category'].unique().tolist(),
    value='All',
    styles=widget_style
)
region_selector = pn.widgets.Select(
    name='Region',
    options=['All'] + df['Region'].unique().tolist(),
    value='All',
    styles=widget_style
)
segment_selector = pn.widgets.Select(
    name='Segment',
    options=['All'] + df['Segment'].unique().tolist(),
    value='All',
    styles=widget_style
)
ship_selector = pn.widgets.Select(
    name='Ship Mode',
    options=['All'] + df['Ship Mode'].unique().tolist(),
    value='All',
    styles=widget_style
)
date_slider = pn.widgets.DateRangeSlider(
    name='Order Date Range',
    start=df['Order Date'].min(),
    end=df['Order Date'].max(),
    value=(df['Order Date'].min(), df['Order Date'].max()),
    sizing_mode='stretch_width',
    styles=widget_style
)

# --- Filter Function ---
def filter_data(df):
    data = df.copy()
    if category_selector.value != 'All':
        data = data[data['Category'] == category_selector.value]
    if region_selector.value != 'All':
        data = data[data['Region'] == region_selector.value]
    if segment_selector.value != 'All':
        data = data[data['Segment'] == segment_selector.value]
    if ship_selector.value != 'All':
        data = data[data['Ship Mode'] == ship_selector.value]
    data = data[(data['Order Date'] >= date_slider.value_start) & (data['Order Date'] <= date_slider.value_end)]
    return data

# --- Charts ---
@pn.depends(category_selector, region_selector, segment_selector, ship_selector, date_slider)
def sales_trend(*events):
    data = filter_data(df)
    sales = data.groupby('Month')['Sales'].sum().reset_index()
    return px.line(sales, x='Month', y='Sales', title='Sales Trend', template='plotly_dark')

@pn.depends(category_selector, region_selector, segment_selector, ship_selector, date_slider)
def profit_margin(*events):
    data = filter_data(df)
    pm = data.groupby('Month').apply(lambda x: (x['Profit'].sum()/x['Sales'].sum())*100).reset_index(name='Profit Margin')
    return px.line(pm, x='Month', y='Profit Margin', title='Profit Margin (%)', template='plotly_dark')

@pn.depends(category_selector, region_selector, segment_selector, ship_selector, date_slider)
def category_sales(*events):
    data = filter_data(df)
    cat_sales = data.groupby('Category')['Sales'].sum().reset_index()
    return px.bar(cat_sales, x='Category', y='Sales', title='Category-wise Sales', template='plotly_dark', color='Category')

@pn.depends(category_selector, region_selector, segment_selector, ship_selector, date_slider)
def region_sales(*events):
    data = filter_data(df)
    reg_sales = data.groupby('Region')['Sales'].sum().reset_index()
    return px.bar(reg_sales, x='Region', y='Sales', title='Region-wise Sales', template='plotly_dark', color='Region')

@pn.depends(category_selector, region_selector, segment_selector, ship_selector, date_slider)
def top_products(*events):
    data = filter_data(df)
    prod_sales = data.groupby('Product Name')['Sales'].sum().nlargest(10).reset_index()
    fig = px.bar(prod_sales, x='Product Name', y='Sales', title='Top 10 Products by Sales', template='plotly_dark')
    fig.update_xaxes(tickangle=45)
    return fig

@pn.depends(category_selector, region_selector, segment_selector, ship_selector, date_slider)
def segment_sales(*events):
    data = filter_data(df)
    seg_sales = data.groupby('Segment')['Sales'].sum().reset_index()
    return px.pie(seg_sales, names='Segment', values='Sales', title='Segment-wise Sales', template='plotly_dark')

@pn.depends(category_selector, region_selector, segment_selector, ship_selector, date_slider)
def ship_mode(*events):
    data = filter_data(df)
    ship_sales = data.groupby('Ship Mode')['Sales'].sum().reset_index()
    return px.pie(ship_sales, names='Ship Mode', values='Sales', title='Sales by Shipping Mode', template='plotly_dark')

# --- Dashboard Layout ---
dashboard = pn.template.MaterialTemplate(
    title="📊 Sales Dashboard (Full Page)",
    theme_toggle=False,
    header_background=bg_color,
    header_color=text_color,
    sidebar=[category_selector, region_selector, segment_selector, ship_selector, date_slider],
    main=[
        pn.Card(sales_trend, title="Sales Trend", background=card_bg_color, sizing_mode="stretch_width"),
        pn.Card(profit_margin, title="Profit Margin", background=card_bg_color, sizing_mode="stretch_width"),
        pn.Card(category_sales, title="Category Sales", background=card_bg_color, sizing_mode="stretch_width"),
        pn.Card(region_sales, title="Region Sales", background=card_bg_color, sizing_mode="stretch_width"),
        pn.Card(top_products, title="Top Products", background=card_bg_color, sizing_mode="stretch_width"),
        pn.Card(segment_sales, title="Segment Sales", background=card_bg_color, sizing_mode="stretch_width"),
        pn.Card(ship_mode, title="Shipping Mode", background=card_bg_color, sizing_mode="stretch_width"),
    ]
)

dashboard.servable()
```
